```python
# .agg() method
def pct30(column):
    return column.quantile(0.3)
```

```python
dogs["Weight_kg"].agg(pct30)
```

```python
dogs[["Weight_kg","Height_cm", "Bmi"]].agg(pct30)
```

```python
# .agg() method
def pct40(column):
    return column.quantile(0.4)
```

```python
dogs["Weight_kg"].agg(pct40)
```

```python
dogs[["Weight_kg","Height_cm", "Bmi"]].agg(pct40)
```

```python
dogs[["Weight_kg","Height_cm", "Bmi"]].agg([pct30, pct40])
```
